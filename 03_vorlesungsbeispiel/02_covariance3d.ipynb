# Kovarianzmatrix einer 3D-Stichprobe 

In diesem Notebook laden wir eine 3-dimensionale Stichprobe und plotten diese als Punktewolke. Wir versuchen dann herauszufinden, was uns die Kovarianzmatrix über die Form dieser Wolke verrät.

Zuerst **laden** wir die Stichprobe (1000 Datenpunkte zu je 3 Dimensionen) aus der Datei *samples.txt*.

```python
import numpy as np

with open("samples.txt") as file:
    lines = file.readlines()
    X = map(lambda x: [float(t) for t in x.split()], lines)
    #print(list(X))
    X = np.array(list(X)) 
    # Gebe die ersten 10 Samples mit print() aus.    
    print(X[:10])
```

Wir **plotten** die Datenwolke. Können wir die Kennwerte der Wolke raten, indem wir die Wolke clever drehen?

```python
%matplotlib notebook

# This import registers the 3D projection, but is otherwise unused.
#from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c='r', marker='o', s=2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

Wir prüfen ob unsere Vermutung richtig war: Wir berechnen mit Numpy den **echten Mittelwert** und die **echte Kovarianzmatrix**.

```python
print('Mean:')
print(X.mean(axis=0))
print
print('Covariance Matrix:')
print(np.cov(X.T))
```
